# Ingestão

Realizar a ingestão dos datasets que estão no banco SQL na Azure em um diretório de arquivos raw dentro do DBFS.

## Configuração para acesso ao banco de dados

Importante!!! Não esquecer de liberar o IP do Databricks nas configurações do banco no Azure.

```
jdbcHostname = "datazilla.database.windows.net"
jdbcDatabase = "datazilla"
jdbcPort = 1433
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2}".format(jdbcHostname, jdbcPort, jdbcDatabase)
connectionProperties = {
  "user" : 'datazilla',
  "password" : 'gama123456@#$',
  "driver" : "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}
```

## Importação das tabelas do banco de dados

```
df_resumo = spark.read.jdbc(url=jdbcUrl, table='SUMARY_PAISES', properties=connectionProperties)
df_pais = spark.read.jdbc(url=jdbcUrl, table='PAIS', properties=connectionProperties)
df_dados_paises = spark.read.jdbc(url=jdbcUrl, table='DADOS_PAISES', properties=connectionProperties)
```

## Salvando os arquivos em json no diretório _raw

```
# Função adicionando data no nome para quando configurada a automatização deste job, seja criado um arquivo por dia

from datetime import date
today=date.today()

df_resumo.write.json("dbfs:/FileStore/_covid_data_lake/_raw/_summary/{}.json".format(today), mode='overwrite')
df_pais.write.json("dbfs:/FileStore/_covid_data_lake/_raw/_pais/{}.json".format(today), mode='overwrite')
df_dados_paises.write.json("dbfs:/FileStore/_covid_data_lake/_raw/_dados_paises/{}.json".format(today), mode='overwrite')
```
